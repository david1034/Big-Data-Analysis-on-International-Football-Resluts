```
#New Data extraction, after data cleaning removing un-needed columns for fitting my model.I took matches,scores from 2000 to 2019 of Brazil.
```

```
#Reading the brazil data into my Data Bricks as a new data
```

```
dataml= sqlContext.sql("SELECT * From brazil_home_score_csv")
dataml.show()
display(dataml)
```

```
# Using the .drop function to drop the date column which was converted in excel to int since this is not a time series regression date column which indicate the date and time of all matches will not be necessary for fittinng our machine learning algorithms since we plan achiving a linear regression.
```

```
dataml=dataml.drop(dataml.date)
display(dataml)
```

```
# Displaying the new data after using the .drop function.
```

```
display(dataml)
```

```
# Importing pipeline and vector assembler 
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
vectorAssembler=VectorAssembler(inputCols=['matches'],outputCol='features')
dataml=vectorAssembler.transform(dataml)
dataml=dataml.select("features","goals")
dataml.show(9)
```

```
# Splitting of data into test and training 
```

```
splits=dataml.randomSplit([0.7,0.3])
train_df=splits[0]
test_df=splits[1]
```

```
# Importing linear regrerssion 
```

```
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('lnr').getOrCreate()
from pyspark.ml.regression import LinearRegression
```

```
# Evaluating linear regression using different parameters and finding the coefficients and intercept of the fitted line
```

```
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol = 'features', 
labelCol='goals', maxIter=10, regParam=0.3, elasticNetParam=0.8)
lr_model=lr.fit(train_df)
print("Coefficients: " + str(lr_model.coefficients))
print("Intercept: " + str(lr_model.intercept))
```

```
# Displaying the train data set 
```

```
train_df.describe().show()
```

```
# Evaluating the train data
```

```
lr_predictions = lr_model.transform(train_df)
lr_predictions.select("prediction","goals","features").show(5)
from pyspark.ml.evaluation import RegressionEvaluator
lr_evaluator = RegressionEvaluator(predictionCol="prediction", \
                 labelCol="score",metricName="r2")
```

```
display(train_df)
```

```
# Sumarry of training
```

```
trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```
